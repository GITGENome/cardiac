```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(18, 18))
count = 0
for col in ['age', 'sex', 'cp', 'trestbps']:
  sns.boxplot(df_cardiac[col], ax = axs[0, count]).set_title(col)
  count+= 1
count = 0
for col in ['chol', 'fbs', 'restecg', 'thalach']:
  sns.boxplot(df_cardiac, x = df_cardiac[col], ax = axs[1, count]).set_title(col)
  count+=1
count=0
for col in ['exang', 'oldpeak', 'slope', 'ca']:
  sns.boxplot(df_cardiac, x = df_cardiac[col], ax = axs[2, count]).set_title(col)
  count+=1

```

```python
def val_coun(col):
  print(f'for {col} there are such values \n{df_cardiac[col].value_counts()}')

for col in ['sex', 'cp', 'fbs', 'oldpeak', 'ca']:
  val_coun(col)
```
